# Data Processing Template

Since almost without exception the first step is working with data munging, this notebook is stubbed out with a very select few imports to speed up the initial steps following creating a project and copying in raw data.

```python
# minimal module imports
from arcgis.features import GeoAccessor, GeoSeriesAccessor
import pandas as pd
import os
import importlib

# import arcpy if available
if importlib.util.find_spec("arcpy") is not None:
    import arcpy
```

```python
# a couple of handy variables and settings to get started

data = r'../data'
data_raw = os.path.join(data, 'raw')
data_ext = os.path.join(data, 'external')
data_int = os.path.join(data, 'interim')
data_out = os.path.join(data, 'processed')

gdb = os.path.join(data_int, 'interim.gdb')
```

```python
coffee_path = os.path.abspath(os.path.join(data_raw, 'raw.gdb\coffee'))
coffee_path
```

```python
coffee_df = GeoAccessor.from_featureclass(coffee_path)
coffee_df.head()
```

```python
df_count = coffee_df[['OBJECTID', 'CONAME']].groupby('CONAME').count()
df_count.columns = ['count']
df_count.sort_values('count', ascending=False, inplace=True)
df_count.reset_index(inplace=True)
df_count
```

```python
mask = df_count['count'] < 3
df_count.loc[mask, 'CONAME'] = 'OTHER'
df_count
```

```python
df_count_sum = df_count.groupby('CONAME')['count'].sum().to_frame()
df_count_sum.sort_values('count', ascending=False, inplace=True)
df_count_sum
```

```python
for idx, (company, count) in df_count_sum.iterrows():
    print(f'{company}{count}')
```
